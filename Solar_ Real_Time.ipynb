```python
url1 = "https://services.swpc.noaa.gov/json/goes/primary/xrays-3-day.json"
url2 = "https://services.swpc.noaa.gov/json/goes/secondary/xrays-3-day.json"


response1 = request.urlopen(url1)
data1 = json.loads(response1.read())

response2 = request.urlopen(url2)
data2 = json.loads(response2.read())
```

```python
df16 = pd.DataFrame.from_dict(data1)
df17 = pd.DataFrame.from_dict(data2)

df16_l = df16[df16.energy != '0.05-0.4nm']
df17_l = df17[df17.energy != '0.05-0.4nm']
df16_l.drop(['satellite','observed_flux', 'electron_correction', 'electron_contaminaton','energy'], inplace=True, axis=1)
df16_l.columns =['time_tag','xrsb_16_l']
df17_l.drop(['satellite','observed_flux', 'electron_correction', 'electron_contaminaton','energy'], inplace=True, axis=1)
df17_l.columns =['time_tag','xrsb_17_l']

df16_s = df16[df16.energy == '0.05-0.4nm']
df17_s = df17[df17.energy == '0.05-0.4nm']
df16_s.drop(['satellite','observed_flux', 'electron_correction', 'electron_contaminaton','energy'], inplace=True, axis=1)
df16_s.columns =['time_tag','xrsb_16_s']
df17_s.drop(['satellite','observed_flux', 'electron_correction', 'electron_contaminaton','energy'], inplace=True, axis=1)
df17_s.columns =['time_tag','xrsb_17_s']

```

```python
df16=df16_l.merge(df16_s,on="time_tag", how="inner")
df17=df17_l.merge(df17_s,on="time_tag", how="inner")
df=df16.merge(df17,on="time_tag", how="inner")
df
```

```python
# Import augmented dicky-fuller test function
from statsmodels.tsa.stattools import adfuller

def check_stationarity(series):
   # Copied from https://machinelearningmastery.com/time-series-data-stationary-python/

    result = adfuller(series.values)

    print('ADF Statistic: %f' % result[0])
    print('p-value: %f' % result[1])
    print('Num of lags: %f' % result[2])
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t%s: %.3f' % (key, value))

    if (result[1] <= 0.05) & (result[4]['5%'] > result[0]):
        print("\u001b[32mStationary\u001b[0m")
    else:
        print("\x1b[31mNon-stationary\x1b[0m")
```

```python
check_stationarity(XRS['xrsb_16'])
```
